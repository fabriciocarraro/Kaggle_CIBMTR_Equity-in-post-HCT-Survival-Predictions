```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
print(df_train.shape)
print(df_test.shape)
df_train.head()
```

```python
df_train.drop(columns=['ID', 'year_hct'], inplace=True)
df_test.drop(columns=['year_hct'], inplace=True)
```

```python
df_train.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

categorical_cols = df_train.select_dtypes(include=['object']).columns

ohe = OneHotEncoder(dtype=int, sparse_output=False)
encoded_train_data = ohe.fit_transform(df_train[categorical_cols])
feature_names = ohe.get_feature_names_out(categorical_cols)

encoded_df_train = pd.DataFrame(
    encoded_train_data, 
    columns=feature_names,
    index=df_train.index
)

# Drop original categorical columns and concatenate encoded ones
df_train_encoded = pd.concat([df_train.drop(columns=categorical_cols), encoded_df_train], axis=1)

pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)

print(df_train_encoded.shape)
df_train_encoded.head(10)
```

```python
import plotly.express as px

fig = px.imshow(df_train_encoded.corr(), text_auto=True, 
                width=2000, height=2000)
fig.show()
```

```python
df_train_encoded.isnull().sum()[df_train_encoded.isnull().sum() > 0]
```

```python
# Fill NaN values with mean for numeric columns that have NaN values
columns_to_fill = ['hla_match_c_high', 'hla_high_res_8', 'hla_low_res_6', 'hla_high_res_6', 
                   'hla_high_res_10', 'hla_match_dqb1_high', 'hla_nmdp_6','hla_match_c_low',
                   'hla_match_drb1_low',  'hla_match_dqb1_low', 'hla_match_a_high', 
                   'hla_match_b_low',  'hla_match_a_low',  'hla_match_b_high',
                   'hla_low_res_8', 'hla_match_drb1_high', 'hla_low_res_10']

df_encoded_filled = df_train_encoded.copy()

for col in columns_to_fill:
    if col in df_train_encoded.columns:
        df_encoded_filled[col] = df_encoded_filled[col].fillna(df_encoded_filled[col].mean())

print(df_encoded_filled.isnull().sum()[df_encoded_filled.isnull().sum() > 0])
```

```python
# Fill remaining NaN values
df_encoded_filled_filled_missing_age_comornity_karnofsky = df_encoded_filled.copy()

df_encoded_filled_filled_missing_age_comornity_karnofsky['donor_age'] = (
    df_encoded_filled_filled_missing_age_comornity_karnofsky['donor_age'].fillna(
        df_encoded_filled_filled_missing_age_comornity_karnofsky['donor_age'].mean()
    )
)
df_encoded_filled_filled_missing_age_comornity_karnofsky['comorbidity_score'] = (
    df_encoded_filled_filled_missing_age_comornity_karnofsky['comorbidity_score'].fillna(
        df_encoded_filled_filled_missing_age_comornity_karnofsky['comorbidity_score'].mean()
    )
)
df_encoded_filled_filled_missing_age_comornity_karnofsky['karnofsky_score'] = (
    df_encoded_filled_filled_missing_age_comornity_karnofsky['karnofsky_score'].fillna(
        df_encoded_filled['karnofsky_score'].mean()
    )
)

# Check shape and remaining NaN values
print(df_encoded_filled_filled_missing_age_comornity_karnofsky.shape)
print(df_encoded_filled_filled_missing_age_comornity_karnofsky.isnull().sum()[df_encoded_filled_filled_missing_age_comornity_karnofsky.isnull().sum() > 1])
```

```python
# For each numeric column, see the variance:
print(df_encoded_filled_filled_missing_age_comornity_karnofsky.var().to_string(float_format="{:.4f}".format))
```

```python
# Or for one-hot columns, see how many are 0 vs. 1:
for col in df_encoded_filled_filled_missing_age_comornity_karnofsky.columns:
    if col != 'donor_age' and col != 'age_at_hct' and col != 'efs_time':
        print(df_encoded_filled_filled_missing_age_comornity_karnofsky[col].value_counts())
```

```python
# Drop ID and columns with low variance
columns_to_drop = ['gvhd_proph_FK+- others(not MMF,MTX)',
                   'dri_score_Missing disease status',
                   'prim_disease_hct_CML',
                   'prim_disease_hct_HD',
                   'tce_imm_match_P/G',
                   'gvhd_proph_CSA +- others(not FK,MMF,MTX)']

df_survival = df_encoded_filled_filled_missing_age_comornity_karnofsky.drop(columns=columns_to_drop, axis=1)
print(df_survival.shape)
```

```python
from lifelines import CoxPHFitter
from lifelines.utils import k_fold_cross_validation
    
SEED = 5
np.random.seed(SEED)

penalizers = [0.001, 0.01, 0.1, 1.0]
best_score = -9999
best_penalizer = None

for p in penalizers:
    cph = CoxPHFitter(penalizer=p)
    scores = k_fold_cross_validation(cph, df_survival, duration_col='efs_time', event_col='efs', k=5, scoring_method='concordance_index')
    avg_score = np.mean(scores)
    print(f"Penalizer={p}, c-index={avg_score:.4f}")

    if avg_score > best_score:
        best_score = avg_score
        best_penalizer = p

print("Best penalizer:", best_penalizer, "with c-index:", best_score)

# Finally, retrain on the entire dataset with your best_value:
best_cph = CoxPHFitter(penalizer=best_penalizer)
best_cph.fit(df_survival, duration_col='efs_time', event_col='efs')
```

```python
categorical_cols_test = df_test.select_dtypes(include=['object']).columns

# Transform the test data with the *already-fitted* encoder
encoded_test_data = ohe.transform(df_test[categorical_cols])

encoded_df_test = pd.DataFrame(
    encoded_test_data, 
    columns=feature_names,   # same feature names from training
    index=df_test.index
)

df_test_encoded = pd.concat(
    [df_test.drop(columns=categorical_cols_test), encoded_df_test], 
    axis=1
)
```

```python
print(df_test_encoded.shape)
df_test_encoded.head()
```

```python
print(df_test_encoded.isnull().sum()[df_test_encoded.isnull().sum() > 0])
```

```python
df_test_encoded_filled = df_test_encoded.copy()

columns_to_fill_test = ['hla_match_c_high', 'hla_high_res_8', 'hla_high_res_10', 'donor_age']

for col in columns_to_fill_test:
    if col in df_test_encoded.columns:
        df_test_encoded_filled[col] = df_test_encoded_filled[col].fillna(df_test_encoded_filled[col].mean())

print(df_test_encoded_filled.isnull().sum()[df_test_encoded_filled.isnull().sum() > 0])
```

```python
print(df_test_encoded_filled.shape)
df_test_encoded_filled.head()
```

```python
df_survival_test = df_test_encoded_filled.drop(columns=columns_to_drop, axis=1)
print(df_survival_test.shape)
```

```python
# Predict partial hazard
cph_risk_scores = best_cph.predict_partial_hazard(df_survival_test)
print(cph_risk_scores)
```

```python
df_final = pd.DataFrame({
    "ID": df_test_encoded_filled["ID"],
    "prediction": cph_risk_scores  # real-valued risk
})

df_final.head()
```
